```python
# Groundwater dataset
groundwater_df = pd.read_csv("../data/raw/groundwater/Groundwater Rielasingen.csv", sep = ';')
groundwater_df
```

```python
groundwater_df = groundwater_df[["Messzeitpunkt", "Messwert"]]
groundwater_df.info()
```

```python
groundwater_df['Messwert'] = groundwater_df['Messwert'].str.replace(',', '.').astype(float)
```

```python
groundwater_df.columns = ['DATE', 'GROUNDWATER_LEVEL']
groundwater_df['DATE'] = pd.to_datetime(groundwater_df['DATE'], format='%d.%m.%Y')
groundwater_df.info()
```

```python
groundwater_df.set_index('DATE', inplace=True)
groundwater_df = groundwater_df.resample('D').ffill() # Using forward fill to repeat the measurements daily until the next one
groundwater_df.reset_index(inplace=True)
groundwater_df
```

```python
df = pd.merge(df, groundwater_df, on='DATE', how='left')
```

```python
df[25000:25100] # Looking at a part where all values exist
```

```python
# Climate change dataset
climate_df = pd.read_csv("../data/raw/climate/Climate Change.csv")
climate_df = climate_df[climate_df["Country"] == "Switzerland"]
climate_df.head()
```

```python
climate_df = climate_df.drop(columns = ["ObjectId","Country","ISO2","ISO3","Indicator","Unit","Source","CTS Code","CTS Name","CTS Full Descriptor"])
climate_df = pd.melt(climate_df, var_name="DATE", value_name="CLIMATE_CHANGE")
climate_df
```

```python
climate_df.info() # No missing values
```

```python
climate_df = pd.concat([climate_df, pd.DataFrame({"DATE": [2024], "CLIMATE_CHANGE": [0.0]})], ignore_index=True) # Adding dummy for forward filling
climate_df['DATE'] = pd.to_datetime(climate_df['DATE'], format='%Y')

climate_df.set_index('DATE', inplace=True)
climate_df = climate_df.resample('D').ffill() # Using forward fill to repeat the measurements daily until the next one
climate_df.reset_index(inplace=True)

climate_df = climate_df[:-1] # Removing dummy
climate_df
```

```python
df = pd.merge(df, climate_df, on='DATE', how='left')
```
